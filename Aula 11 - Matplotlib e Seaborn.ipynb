```python
# Permite a geração de gráficos
import matplotlib.pyplot as plt

# Permite trabalhar com matrizes e vetores
import numpy as np
```

Primeiro Gráfico com Mtplolib

```python
# Criando os vetores
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Plotar o Gráfico
plt.plot(x,y)
```

#### Modificando os tipos de marcadores

```python
plt.plot(x,y, marker= '*')
```

### Alterando as cores da linha

```python
plt.plot(x,y, marker= '*', c='g')
```

```python
plt.plot(
    x,y, # Os dados do Gráfico
    marker='*', # Tipo de maercador
    c ='green', # Corda linha
    ds ='steps', # A forma com que os pontos são conectados
    ls ='--', # Estilo da linha
    mec ='red' # Cor dos marcadores
)
```

```python
plt.plot(x,y)

# adicionar titulo ao eixo x
plt.xlabel('valor de X', fontsize = '12')

# adicionar titulo ao eixo x
plt.ylabel('valor de y', fontsize = '12', c='green')

# adicionar titulo ao grafico
plt.title('Titulo do grafico', fontsize = '12', c='purple')

# Altera as cores dos valores dos eixos
plt.xticks(fontsize= 8, c='blue')

plt.yticks(fontsize = 8, c='green')
```
