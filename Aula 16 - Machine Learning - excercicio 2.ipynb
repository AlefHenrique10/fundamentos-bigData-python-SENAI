```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# mapeamento do caminho e arquivo 
path = 'C:\\Users\\FIC\\Documents\\Henrique-Rodrigues\\Aquivo atualizado\\017 - Python - Fundamentos de Big Data e Data Analytics\\002 - Exercícios\\Machine Learning-20231006T223445Z-001\\Machine Learning\\'
file = 'protetor_solar.csv'

# Atribuir o arquivo a um dataframe
protetor_Solar = pd.read_csv(path + file)

# Verificar o cabeçalho
protetor_Solar.head()
```

```python
### Grafico de dispersão e histograma
sns.jointplot(x = 'temperatura', y = 'qtde', data = protetor_Solar, color = 'green')
```

```python
sns.pairplot(protetor_Solar)
```

```python
sns.lmplot(x = 'temperatura',
           y = 'qtde',
           data = protetor_Solar)
```

```python
x = protetor_Solar[['temperatura']]
y = protetor_Solar['qtde']
```

```python
# importando o comando train_test_split para dividir o DataFrame
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train,y_train)
```

```python
print('Coeficiente angular (a):', regressor.coef_)
print('Coeficiente linear (b):', regressor.intercept_)
```

```python
y_previsao = regressor.predict(X_test)

y_previsao
```

```python
plt.scatter(X_train,y_train, color = 'green')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Quantidade')
plt.xlabel('Temperatura ')
```
