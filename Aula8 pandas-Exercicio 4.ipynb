a) Crie um dataframe, utilizando o arquivo dados-vendas.xlsx. Crie data frames
para dividir a planilha vendas em 4 arquivos. São eles:
Tabela Fato

- ID-Produto;

- ID-Loja;

- ID-Vendedor;

- Data;

- ValorVenda.

- Produto;

Tabelas Vendedor

```python
# Importar a biblioteca do pandas
import pandas as pd

# Mapear o local do documento fonte

caminho = 'C:\\Users\\FIC\\Documents\\Henrique-Rodrigues\\003 - Arquivos\\'
arquivo = 'dados-vendas.xlsx'

# Atribuir os dados deste arquivo, em um dataframe
base = pd.read_excel(caminho + arquivo, sheet_name='Vendas')
base = base.rename(columns={'Loja': 'ID-Loja'})


# Exibir os cabeçalhos das colunas

base.head()
```

```python
# Criando a tabela fato com as coluna:

df_dados= base[['ID-Produto', 'ID-Loja', 'ID-Vendedor', 'Data Venda', 'ValorVenda']]

# renomear a coluna Loja para ID-Loja
df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

df_dados.head()

# Slavar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'Dados - venda\\'
arquivo = 'tabela_fato.xlsx'
df_dados = df_dados.drop_duplicates()

df_dados.to_excel(caminho_novo + arquivo, index=False)
```

```python
tabela_produtos = base[['ID-Produto','Produto', 'Categoria','Segmento','Fabricante']]

caminho_novo = caminho + 'Dados - venda\\'
arquivo = 'Tabela_Produtos.xlsx'
tabela_produtos = tabela_produtos.drop_duplicates()

tabela_produtos.to_excel(caminho_novo + arquivo, index=False)
tabela_produtos.head()
```

```python
tabela_Loja = base[['ID-Loja','Cidade','Estado']]

caminho_novo = caminho + 'Dados - venda\\'
arquivo = 'Tabela_Loja.xlsx'
tabela_Loja = tabela_Loja.drop_duplicates()

tabela_Loja.to_excel(caminho_novo + arquivo, index=False)
tabela_Loja.head()
```

```python
tabela_Vendedor = base[['ID-Vendedor','Vendedor']]

caminho_novo = caminho + 'Dados - venda\\'
arquivo = 'Tabela_Vendedor.xlsx'
tabela_Vendedor= tabela_Vendedor.drop_duplicates()

tabela_Vendedor.to_excel(caminho_novo + arquivo, index=False)

tabela_Vendedor.head()
```
