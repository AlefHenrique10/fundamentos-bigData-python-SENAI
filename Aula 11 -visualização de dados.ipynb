Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

```python
plt.plot(x,y, label = 'seno')
plt.plot (x,z,label='cosseno')
plt.plot (x,np.sqrt(x), label='raiz quadrada')
plt.legend()
```

Matplotlib – Múltiplos subgráficos

```python
# O comando plt.subplot permir criar grupo de gráfico
# plt.subplot (linhas, colunas, posição elemento)

plt.subplot(2,1,1) # Primeiro gráfico
plt.plot(x,y) # Conjunto de dados do primeiro 
plt.title('subplots') 
plt.subplot(2,1,2) # Segundo gráfico
plt.plot(x,z) # Conjunto de dados do segundo gráfico
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```

### Seaborn – Gráficos mais elaborados

Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos, 
além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.

Importando a biblioteca

```python
# O ALIAS ( apelido) da biblioteca SEABORN dica como sns
import seaborn as sns
import numpy as np

# Retorna ( Linspace ) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)

```

Criando um dataframe com o dataset íris.

```python
iris = sns.load_dataset('iris')
iris.head()
```

Criando um dataframe com o dataset planets.

```python
planetas = sns.load_dataset('planets')
planetas.head()
```

## Seaborn – Ajustando os gráficos

O Seaborn possui a função set () que permite ajustar estilos específicos para os gráficos, sem a 
necessidade de ajustar manualmente todos os argumentos.

```python

plt.plot (x,y)
```

```python
# alterando o tema para claro
sns.set_style('whitegrid')
plt.plot(x,y)
```

```python
# alterando o tema para claro
sns.set_style('dark')
plt.plot(x,y)
```
